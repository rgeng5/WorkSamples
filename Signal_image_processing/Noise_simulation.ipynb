```python
import numpy as np
import matplotlib.pyplot as plt
# Parameters of Rician noise
v = 8
s = 5
N = 100  # how many samples

noise = np.random.normal(scale=s, size=(N, 2)) + [[v,0]]
noise = np.linalg.norm(noise, axis=1)
noise = noise.reshape((10,10))
plt.imshow(noise)
plt.colorbar()
```

```python
from PIL import Image
image = np.array(Image.open('s5_b1_r1.jpg').convert('L'))

s_vs_p = 0.5
a = [0.03,0.08,0.15]

for amount in a:
  out = np.copy(image)
  # Salt mode
  num_salt = np.ceil(amount * image.size * s_vs_p)
  coords = [np.random.randint(0, i - 1, int(num_salt))
          for i in image.shape]
  out[coords] = 255

  # Pepper mode
  num_pepper = np.ceil(amount* image.size * (1. - s_vs_p))
  coords = [np.random.randint(0, i - 1, int(num_pepper))
          for i in image.shape]
  out[coords] = 0

  plt.imshow(out, cmap='gray', vmin=0, vmax=85)
  plt.axis('off')
  plt.show()
```
